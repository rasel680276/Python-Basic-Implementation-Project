```python
# 15
def prime(x):
    if x<2:
        return 'not prime'
    for i in range(2,x):
        if x%i==0:
            return 'not prime'
    return "prime"
a = int(input())
print(prime(a))
```

```python
# 16
def lv(x):
    v = ['a','e','i','o','u']
    x = x.lower()
    f = 0
    g = ""
    for i in v:
        c = x.count(i)
        if c > f:
            f = c
            g = i
    return "max is ", g
inpu = input()
print(lv(inpu))
```
